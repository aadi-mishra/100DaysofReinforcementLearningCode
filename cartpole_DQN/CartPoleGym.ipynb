CartPole is a classical control environment provided by Gym and used by researchers as a starting point of algorithms. It consists of a cart that moves along the horizontal axis (1-dimensional) and a pole anchored to the cart on one endpoint.
The agent has to learn how to move the cart to balance the pole

State Space is represented by position and velocity of the cart and angular position and angular veocity of the pole. The state space is continous in this case but can be discretized to make our problem simpler.

```python
import gym
from typing import Tuple
from gym import spaces
```

```python
env = gym.make('CartPole-v1')

print("Action Space:", env.action_space)
print("Observation Space", env.observation_space)
```

```python
# Analyzing the observation space 
print("Lower bound of the Observation Space:", env.observation_space.low)
print("Upper bound of the Observation Space:", env.observation_space.high)
```

```python
env.action_space.sample()
```

```python
# Create a discrete space composed by N-elements (5)
n:int = 5
discrete_space = spaces.Discrete(n=n)
discrete_space.seed(21)

# Sample from the space using .sample method
print("Discrete Space Sample:", discrete_space.sample())

box_shape: Tuple[int, int] = (4, 4)
box_space = spaces.Box(low=0, high=1, shape=box_shape)
box_space.seed(21)
print("Box Space Sample:", box_space.sample())
```

### Creating a space for Image Observations

```python
from gym import spaces
import matplotlib.pyplot as plt
import numpy as np
```

```python
shape = (256, 256, 3)
low = 0
high = 255
dtype = np.uint8
space = spaces.Box(low=low, high=high, shape=shape, dtype=dtype)
print("Space", space)
```

```python
sample = space.sample()
print("Space Sample", sample)
```

```python
plt.imshow(sample)
```

```python
n_frames = 4
width = 256
height = 256
channels = 3
shape_temporal = (n_frames, width, height, channels)
space_temporal = spaces.Box(low=low, high=high, shape=shape_temporal, dtype=dtype)
print("Space with temporal component", space_temporal)
```

```python
env = gym.make("CartPole-v1")
env.reset()
n_steps = 100

for i in range(n_steps):
  action = env.action_space.sample()
  env.step(action)
  env.render()
# close the environment correctly
  env.close()
```
